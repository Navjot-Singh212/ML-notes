**Machine Learning** is the field of study that gives computers the ability to learn without being explicitly programmed.

Machine Learning algorithms:

-Supervised Learning

-Unsupervised Learning

-Recommender systems

-Reinforcement learning

***Supervised learning***
it is a ml algorithm in which the model learns from being given the "right answers". *basically* we give the model the input **X** and the correct output **Y**.

Basically the supervised leanring is furthur divided into two types based on the predicted output Y' it gives:

**1. Regression**: Predicts a number from infintiely many possibile outputs.

**2. Classification:** Predicts a category from small number of possibile outputs.

***Unsupervised Learning:*** it is also a ml algorithm in which the algorihtm finds something interesting or some pattern or a structure in the unlabeled data.
Basically in this we feed it unlabeled data i.e., the data comes with only the inputs **X** but not the output **Y**.
The algorithm has to find a structure in the data on its own.

It is also further divided into three parts based on what a specific algorithm does:

**1. Clustering:** Groups similar objects together.

**2. Anomaly Detection:** Find unusual data points.

**3.Dimensionality reduction:** Compress data using fewer numbers.

(we will learn more about anomaly detection and dimensionality reduction later in the course.)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1. The Data
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

# 2. The Function (The "Model")
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
# Change these numbers to move the line!
w = 200
b = 100

# Calculate the line
tmp_f_wb = compute_model_output(x_train, w, b)

# Plotting
plt.figure(figsize=(6,4)) # Optional: Makes the graph size nice
plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
